```python
def is_float(x):
        try:
            float(x)
        except:
            return False
        return True
```

```python
df5[~df5['Rooms'].apply(is_float)].head(100)
```

```python
def convert_rooms_to_bhk(x):
    tokens = x.split('+')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))
    try:
        return float(x)
    except:
        return None    
```

```python
def convert_rooms_to_bhk2(x):
    tokens = x.split('+')
    if len(tokens)==2:
        return (float(tokens[0])+ 2 )
    try:
        return float(x)
    except:
        return None    
```

```python
convert_rooms_to_bhk('3')
```
