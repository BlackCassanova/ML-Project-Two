```python
def is_float(x):
        try:
            float(x)
        except:
            return False
        return True
```

```python
df6[~df6['Rooms'].apply(is_float)].head()
```

```python
def convert_rooms_to_bhk(x):
    tokens = x.split('+')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))
    try:
        return float(x)
    except:
        return None 
```

```python
convert_rooms_to_bhk('3+1')
```

```python
convert_rooms_to_bhk('3')
```

```python
def plot_scatter_chart(df, lo):
        bhk2 = df[(df['Location'].str.contains(lo)) & (df['Rooms']==2)]
        bhk3 = df[(df['Location'].str.contains(lo)) & (df['Rooms']==3)]
        bhk4 = df[(df['Location'].str.contains(lo)) & (df['Rooms']==4)]
        
        matplotlib.rcParams['figure.figsize'] = (15,10)
        
        plt.scatter(bhk2['Size'], bhk2['Price'], color = 'red', label='2 BHK',  marker="1", s=50)
        plt.scatter(bhk3['Size'], bhk3['Price'], color = 'blue', label='3 BHK',  marker="2", s=50)
        plt.scatter(bhk4['Size'], bhk4['Price'], color = 'green', label='4 BHK',  marker="3",s=50)
        
        plt.xlabel("Size sqft")
        plt.ylabel("Price Lakh RM")
        plt.title(lo)
        plt.legend()
```

```python
plot_scatter_chart(adf8,"Mont Kiara")
```
